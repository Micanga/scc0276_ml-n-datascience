### Questão 02.

Nessa questão, você irá fazer o pré-processamento dos dados.

Você deve:


*   **Para os dados numéricos:** substitua os valores faltantes utilizando a estratégia de média (`sklearn.impute.SimpleImputer`). Depois padronize o intervalo dessas variáveis (`sklearn.preprocessing.StandardScaler`)
*   **Para os dados categóricos:** substitua os valores faltantes utilizando a estratégia de mais frequentes (`sklearn.impute.SimpleImputer`). Depois converta os dados categóricos para uma representação numérica (`sklearn.preprocessing.OneHotEncoder`),

Lembre-se de utilizar as classes `sklearn.pipeline.Pipeline` e `sklearn.compose.ColumnTransformer` conforme consta no material complementar. 

**RESPOSTA:**

Primeiramente, iremos analisar os formatos dos parâmetros e dividir o conjunto de dados.

```python
dados.info()
```

Utilizando as informações disponibilizadas pelo método "info", dividiremos os dados por seus tipos.
Isto é, dividiremos os dados entre os grupos de "parâmetros numéricos" e "parâmetros categóricos".
Trivialmente, a coluna classe não será pré-processada.

```python
features_numericos =  ['AGE','BILIRUBIN','ALK_PHOSPHATE','SGOT','ALBUMIN','PROTIME']
features_categoricos = ['SEX','STEROID','ANTIVIRALS','FATIGUE','MALAISE','ANOREXIA',\
    'LIVER_BIG','LIVER_FIRM','SPLEEN_PALPABLE','SPIDERS','ASCITES','VARICES','HISTOLOGY']
```

Realizada a separação dos dados, partiremos para a criação dos Pipelines de transformação como definido pelo exercício.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

pipeline_numerico = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

pipeline_categorico = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())])
```

Com os Pipelines definidos, partiremos para a aplicação deste sobre as colunas.

```python
from sklearn.compose import ColumnTransformer
transformacao = ColumnTransformer(
    transformers=[
        ('transformacao numerica', pipeline_numerico, features_numericos),
        ('transformacao categorica', pipeline_categorico, features_categoricos),        
    ])

dados_transformados = transformacao.fit_transform(dados)
dados_transformados_com_classe = np.c_[dados_transformados, dados['Class']]
dataframe_processado = pd.DataFrame(dados_transformados_com_classe)
dataframe_processado
```

Feita a transformação, o detalhe faltante para completar o conjunto de dados é caracterizar as colunas novamente.
A ordem será correspondente a ordem de pré-processamento dos parâmetros aplicados (adicionando a classe ao final).

```python
nomes = []

# adicionando as features numericas
nomes = nomes + features_numericos

# adicionando as features categoricas
transformacao_categorica = transformacao.transformers_[1]
pipeline_categorico = transformacao_categorica[1]
transf_onehot = pipeline_categorico.named_steps['onehot']
nomes = nomes + (transf_onehot.get_feature_names().tolist())

# adicionando a classe
nomes.append("classe")

dataframe_processado = pd.DataFrame(data = dados_transformados_com_classe, columns=nomes)
dataframe_processado
```

### Questão 03.

Determine as 10 variáveis mais relevantes do conjunto (lembre-se de concatenar a variável alvo no dataset transformado). Você pode utilizar a matriz de covariância (`numpy.cov`), correlação (`pandas.DataFrame.corr`), `scatterplot`, etc. Justifique sua escolha.

Lembre-se que heatmaps (`seaborn.heatmap`) podem auxiliar a visualização.

```python
import matplotlib.pyplot as plt
import numpy as np
from seaborn import heatmap

# Coletando os parâmetros e eliminando a classe
X = (dataframe_processado.sort_values(by='classe',axis=0)).iloc[:,:-1]

# Criando a figura
f, axs = plt.subplots(3, 1,figsize=(16, 32))

# Criando o Heatmap de Covariância
heatmap(np.cov(X), ax=axs[0], cmap=plt.cm.Blues)

# Criando o Heatmap de Correlação
heatmap(X.corr(method= 'pearson'), ax=axs[1], cmap=plt.cm.Blues)

# Criando um scatterplot dos dados
x = [np.arange(len(X.iloc[i])) for i in range(len(X))]
y = X
colors = np.array([dataframe_processado.iloc[i,-1] for i in range(len(x)) for j in range(len(x[i]))])
colors.resize((np.array(y).shape[0],np.array(y).shape[1]))
axs[2].scatter(x,y,c=colors)

# Plotando
plt.show()
```

Pela criação dos respectivos gráficos, não foi possível se visualizar quais as variáveis mais relevantes no conjunto de dados.

Desta forma, optou-se pela aplicação do método PCA para extração dessa informação.

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X)

componentes_principais = []
for clist in pca.components_:
    if list(clist).index(max(clist)) not in componentes_principais:
        componentes_principais.append(list(clist).index(max(clist)))

componentes_principais_nome = [nomes[c] for c in componentes_principais]
componentes_principais_nome = componentes_principais_nome[:10]
print('| | componentes escolhidas (índice):',componentes_principais[:10])
print('| | componentes escolhidas (nome):',componentes_principais_nome)
```

### Questão 04.

Com as 10 varíaveis determinadas na questão anterior, crie um novo conjunto de dados.

Lembre-se que o numpy permite indexar colunas/linhas utilizando arrays.

```python
headers = componentes_principais_nome
headers.append('classe')
novo_dataframe = dataframe_processado[headers]
novo_dataframe
```

### Questão 05.

Reduza a dimensionalidade do conjunto de dados criado utilizando a técnica PCA (`sklearn.decomposition.PCA`). Reduza a dimensionalidade de forma que no mínimo 80% da variância dos dados seja mantida (atente-se ao atributo  `explained_variance_ratio_`). Depois exiba os eixos principais.

```python
pca_80 = PCA(.80)
pca_80.fit(novo_dataframe)

componentes_principais_80 = []
for clist in pca.components_:
    if list(clist).index(max(clist)) not in componentes_principais:
        componentes_principais_80.append(list(clist).index(max(clist)))

componentes_principais_80_nome = [nomes[c] for c in componentes_principais] 
componentes_principais_80_nome = componentes_principais_nome[:pca_80.n_components_]
print('Variância:',pca_80.explained_variance_ratio_)
print('Eixos principais:',componentes_principais_80_nome)
```

### Questão 06.

Teste um modelo de classificação (através da função `model_tester`) nos conjuntos criados nas questões 4 e 5. Houve redução expressiva da acurácia?

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA

# - Descrição:  Função que testa um modelo de classificação (MLP). 
# - Parâmetros: Variáveis de entrada (inputs) e a variável alvo (target).
# - Retorna:    Não retorna nenhum valor. Mas imprime a acurácia do modelo.
def model_tester(inputs, target):
  clf = MLPClassifier(hidden_layer_sizes=(5,), random_state=0, max_iter=10000)
  scores = cross_val_score(clf, inputs, target, cv=10)
  print("Score do classificador: %.2f" % (scores.mean()*100))
```

```python
X_10componentes, y_10componentes = novo_dataframe[componentes_principais_nome],novo_dataframe['classe']
model_tester(X_10componentes,y_10componentes)

X_80porcento, y_80porcento = novo_dataframe[componentes_principais_80_nome],novo_dataframe['classe']
model_tester(X_80porcento,y_80porcento)
```

**RESPOSTA:**

Sim, houve uma redução expressiva no "score" do classificador.
Provavelmente, com uma redução de informação para 80% sobre o conjunto de dados original traria melhores resultados do que a redução sobre o conjunto já reduzido para 10 variáveis (aumentado o impacto da parda de informação).
